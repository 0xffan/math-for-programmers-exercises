# 二维向量绘图

```python
from vector_drawing import *

dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4), 
                (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), 
                (0,-3), (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)]

draw(
  Points(*dino_vectors),
  Polygon(*dino_vectors)
)
```

练习2.5：当坐标在-10到10的范围内时，使用draw函数绘制表示向量`(x, x**2)`的点。

```python
ex2_5_points = [(x, x**2) for x in range(-10,11)]
draw(
    Points(*ex2_5_points),
    grid=(1, 10),
    nice_aspect_ratio=False
)
```

向量加法

```python
from vectors import *

dino_vectors2 = [v_add(v, (-1.5,-2.5)) for v in dino_vectors]

draw(
    Points(*dino_vectors),
    Polygon(*dino_vectors),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red)
)
```
